```python
import pandas as pd
df=pd.read_csv("/content/preprocessing_practice_dataset.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Experience_Years'']
df['Monthly_Salary']=df['Monthly_Salary'].fillna(df['Monthly_Salary'].mean())
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
df
```

df.isnull()

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.barplot(data=df,x="Age",y="Gender",palette='pastel')
plt.title("Age vs Gender")
plt.xlabel("Age")
plt.ylabel("Gender")
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))

sns.boxplot(data=df,x="Department",y="Experience_Years",palette='pastel')
plt.title("Department vs Experience")
plt.xlabel("Department")
plt.ylabel("Experience")
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.lineplot(data=df,x="City",y="Gender",palette='pastel')
plt.title("City vs Gender")
plt.xlabel("City")
plt.ylabel("Gender")
#
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.countplot(data=df,x="Department",palette='pastel')
plt.title("Department Distribution")
plt.xlabel("Department")
plt.ylabel("Count")
#
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.histplot(data=df,x="Monthly_Salary",palette='pastel')
plt.title("Monthly Salary Distribution")
plt.xlabel("Monthly Salary")
plt.ylabel("Frequency")
#
```

```python
from sklearn.preprocessing import LabelEncoder
encoded_data=df.copy()
one_hot_encoded=pd.get_dummies(encoded_data['Department'],prefix="Department")
encoded_data=pd.concat([encoded_data,one_hot_encoded],axis=1)
encoded_data
```

```python
performance_class={1:"one",2:"two",3:"three",4:"four",5:"five"}
encoded_data['Performance_Class']=encoded_data['Performance_Rating'].map(performance_class)
encoded_data
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
min_max_scaler=MinMaxScaler()
standard_scaler=StandardScaler()
encoded_data['Min_Max_Age']=min_max_scaler.fit_transform(encoded_data[['Age']])
encoded_data['standard_Salary']=standard_scaler.fit_transform(encoded_data[['Monthly_Salary']])
encoded_data
```
